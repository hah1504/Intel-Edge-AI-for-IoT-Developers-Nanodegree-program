```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# FP32 plots

```python
cpu_fp32 = pd.read_csv("resources/hardware_runs/FP32_stats/cpu_stats.txt", header=None)
cpu_fp32.columns = ['metric', 'value (s)']
cpu_fp32['device'] = 'i5 CPU'
```

```python
plt.figure(figsize=(15,5))
sns.barplot(y="metric", x='value (s)', data=cpu_fp32).set_title("FP32 i5 CPU Stats")
plt.savefig("resources/hardware_runs/FP32_stats/fp32_cpu_stats.png")
```

```python
gpu_fp32 = pd.read_csv("resources/hardware_runs/FP32_stats/gpu_stats.txt", header=None)
gpu_fp32.columns = ['metric', 'value (s)']
gpu_fp32['device'] = 'GPU'
```

```python
plt.figure(figsize=(15,5))
sns.barplot(y="metric", x='value (s)', data=gpu_fp32).set_title("FP32 GPU Stats")
plt.savefig("resources/hardware_runs/FP32_stats/fp32_gpu_stats.png")
```

```python
fpga_fp32 = pd.read_csv("resources/hardware_runs/FP32_stats/fpga_stats.txt", header=None)
fpga_fp32.columns = ['metric', 'value (s)']
fpga_fp32['device'] = 'FPGA'
```

```python
plt.figure(figsize=(15,5))
sns.barplot(y="metric", x='value (s)', data=fpga_fp32).set_title("FP32 FPGA Stats")
plt.savefig("resources/hardware_runs/FP32_stats/fp32_fpga_stats.png")
```

```python
xeon_fp32 = pd.read_csv("resources/hardware_runs/FP32_stats/xeon_cpu_stats.txt", header=None)
xeon_fp32.columns = ['metric', 'value (s)']
xeon_fp32['device'] = 'Xeon CPU'
```

```python
plt.figure(figsize=(15,5))
sns.barplot(y="metric", x='value (s)', data=xeon_fp32).set_title("FP32 Xeon CPU Stats")
plt.savefig("resources/hardware_runs/FP32_stats/fp32_xeon_stats.png")
```

```python
atom_fp32 = pd.read_csv("resources/hardware_runs/FP32_stats/atom_cpu_stats.txt", header=None)
atom_fp32.columns = ['metric', 'value (s)']
atom_fp32['device'] = 'Atom CPU'
```

```python
plt.figure(figsize=(15,5))
sns.barplot(y="metric", x='value (s)', data=atom_fp32).set_title("FP32 Atom CPU Stats")
plt.savefig("resources/hardware_runs/FP32_stats/fp32_atom_stats.png")
```

```python
fp32 = pd.concat([cpu_fp32, gpu_fp32, fpga_fp32, xeon_fp32, atom_fp32])

plt.figure(figsize=(15,5))
sns.barplot(y="metric", x='value (s)', hue='device', data=fp32).set_title("FP32 Stats")
plt.savefig("resources/hardware_runs/FP32_stats/fp32_stats.png")
```
